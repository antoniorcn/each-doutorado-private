```python
modelos = [modelo8, modelo9, modelo10, modelo11]
epocas_train = []
epocas_validation = []
train_loss_norm = []
train_acc_norm = []
train_f1_norm = []
val_loss_norm = []
for index_modelo, modelo in enumerate(modelos):
    qtd_epocas = len(modelo["training_loss"])
    print("Epocas: ", qtd_epocas)
    epocas_train.append(np.arange(1, qtd_epocas + 1))
    epocas_validation.append(np.arange(1, qtd_epocas + 2))
    train_loss = np.array(modelo["training_loss"])
    train_acc = np.array(modelo["acc"])
    train_f1 = np.array(modelo["f1"])
    val_loss = np.array(modelo["validation_loss"])

    train_loss_norm.append((train_loss - train_loss.min()) / (train_loss.max() - train_loss.min()))
    train_acc_norm.append((train_acc - train_acc.min()) / (train_acc.max() - train_acc.min()))
    train_f1_norm.append((train_f1 - train_f1.min()) / (train_f1.max() - train_f1.min()))
    val_loss_norm.append((val_loss - val_loss.min()) / (val_loss.max() - val_loss.min()))
```
